```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
%matplotlib inline
```

```
# 創建常態分佈
samples = np.random.normal(loc=5, size=100000) #常態分布中心位於5，給100000個點
# 過濾掉小於0的值
samples = samples[samples >= 0]
# 計算偏態
skewness = skew(samples) # skew可算偏態，擁有很多統計功能涵式
plt.title(f"Skewness:{skewness}")
sns.distplot(samples)
```

```
# 大於一次方
samples_2 = np.power(samples, 2)
skewness = skew(samples_2) # skew可算偏態，擁有很多統計功能涵式
plt.title(f"Skewness:{skewness}")
sns.distplot(samples_2)
```

```
# 指數成長
samples_2 = np.power(2, samples)
skewness = skew(samples_2) # skew可算偏態，擁有很多統計功能涵式
plt.title(f"Skewness:{skewness}")
sns.distplot(samples_2)
```

```
# 小於一次方，值小的時候，左偏的影響沒有那麼劇烈
samples_2 = np.power(samples, 1/2)
skewness = skew(samples_2) # skew可算偏態，擁有很多統計功能涵式
plt.title(f"Skewness:{skewness}")
sns.distplot(samples_2)
```
